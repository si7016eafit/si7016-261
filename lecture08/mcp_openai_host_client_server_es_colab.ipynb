
# Ejemplo sencillo de MCP con OpenAI: Host, Client y Server

Este notebook para **Google Colab** muestra un ejemplo simple de las entidades principales del **Model Context Protocol (MCP)**:

- **MCP Host**
- **MCP Client**
- **MCP Server**

Usaremos un enfoque práctico con **OpenAI** y un **MCP Server remoto**.

---

## Idea general

En este ejemplo:

- **Host**: tu aplicación Python / notebook
- **Client**: la conexión MCP que se establece internamente a través de OpenAI
- **Server**: un servidor MCP remoto que expone contexto o herramientas

---

## Flujo conceptual

```text
Usuario
  ↓
MCP Host (tu app)
  ↓
MCP Client (conexión MCP)
  ↓
MCP Server (servidor remoto)
  ↓
Tools / contexto / documentación
```


## 3. Importar el cliente de OpenAI


````python
from openai import OpenAI

client = OpenAI()
print("Cliente OpenAI inicializado.")
````


## 4. Explicación de las entidades

### MCP Host
Es la aplicación principal con la que interactúa el usuario.  
En este notebook, el **host** es este mismo código Python.

### MCP Client
Es el componente que mantiene la conexión entre el host y el servidor MCP.  
En este ejemplo, el cliente MCP **no lo programamos manualmente**: la integración MCP se usa a través de la API de OpenAI.

### MCP Server
Es el servicio remoto que expone herramientas, contexto o recursos.
En este notebook lo representaremos como una URL de servidor MCP.



## 5. Definir el servidor MCP remoto

Aquí definimos los metadatos del servidor MCP que queremos usar.

> Puedes reemplazar la URL por el servidor MCP que desees usar en tus propios experimentos.


````python
mcp_server = {
    "type": "mcp",
    "server_label": "demo_mcp_server",
    "server_description": "Servidor MCP remoto para demostración",
    "server_url": "https://developers.openai.com/mcp",
    "require_approval": "never"
}

mcp_server
````


## 6. Ejemplo mínimo: Host + Client + Server

En esta celda, el **Host** (tu notebook) hace una solicitud al modelo de OpenAI y le permite usar un **servidor MCP remoto** como tool/contexto externo.

### Qué ocurre conceptualmente:
1. El usuario hace una pregunta.
2. El Host la envía al modelo.
3. El modelo tiene acceso a una tool de tipo `mcp`.
4. Esa tool representa la conexión al servidor MCP.
5. El modelo consulta el servidor si lo necesita.
6. El Host recibe la respuesta final.


````python
pregunta = (
    "Explica de forma sencilla cuáles son las diferencias entre "
    "MCP Host, MCP Client y MCP Server."
)

respuesta = client.responses.create(
    model="gpt-4.1-mini",
    tools=[mcp_server],
    input=pregunta
)

print(respuesta.output_text)
````


## 7. Interpretación del ejemplo

### En este notebook:
- **Host** = este notebook / tu aplicación Python
- **Client** = la conexión MCP gestionada por la integración de OpenAI
- **Server** = el servidor remoto definido en `server_url`

### Punto importante
No necesitas implementar manualmente el protocolo completo para empezar a experimentar con MCP en este patrón simple.


## 9. Variante: encapsular el Host en una función

Esto permite ver más claramente el papel del **Host** como orquestador.


````python

def ejecutar_host_mcp(pregunta_usuario: str):
    respuesta = client.responses.create(
        model="gpt-4.1-mini",
        tools=[mcp_server],
        input=pregunta_usuario
    )
    return respuesta.output_text

salida = ejecutar_host_mcp(
    "Describe un ejemplo simple de interacción entre un MCP Host, un MCP Client y un MCP Server."
)

print(salida)
````
